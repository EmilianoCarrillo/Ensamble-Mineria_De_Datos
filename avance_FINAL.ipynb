## 1. Preprocesar archivo de texto

```python
nombre_archivo = input("Nombre del Archivo con Datos [cre300, Slice409]:")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import warnings
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
warnings.filterwarnings('ignore')
%matplotlib inline

## **************************************

try:
    with open( "data/" + nombre_archivo + ".txt", "r") as archivo:
        nElem = int(archivo.readline())
        nAtrib = int(archivo.readline())
        nClases = int(archivo.readline())
        
        atributos = []
        for i in range(0, nAtrib):
            atributos.append("atrib_" + str(i+1))
        
        atributos.append("clase")
        data = pd.read_csv(archivo, delimiter=',', header=None)
        data.columns = atributos
    
except FileNotFoundError:
    print( "ERROR: El archivo " + nombre_archivo + " no fue encontrado");
finally:
    archivo.close();

dataset = data;

## **************************************
x_fin = dataset.drop('clase', 1)
y_fin = dataset['clase']
nComponentes = "Sin Componentes"
```

#### Preprocesar datos (Estandarizar)

```python
x = dataset.drop('clase', 1)
y = dataset['clase']

x_estandarizada = StandardScaler().fit_transform(x)

try:
    atributos.remove('clase')
except:
    print('')
    
x_fin = pd.DataFrame(data = x_estandarizada, columns = atributos)
```

## 2. Input del usuario

```python
in_n_components = int(input('PCA \t|| Cantidad de Componentes Principales: '))
in_splits = int(input("\nkFold \t|| Cantidad de pliegues: "))

in_n_neighbors = int(input("\nKNN \t|| Cantidad de Vecinos: "))
in_max_iter = int(input("\nNN \t|| Cantidad de Iteraciones: "))
in_momentum = float(input("   \t|| Momentum (de 0 a 1): "))
in_nHidden_layers = int(input("   \t|| Cantidad de capas ocultas: "))

in_hidden_layer_sizes=[]
for i in range(1,in_nHidden_layers+1):
    in_hidden_layer_sizes.append(int(input("   \t|| | Pereptrones de la capa "+str(i)+": ")))
```

#####  3. PCA

```python
pca = PCA(n_components= in_n_components)
atributos = []
for i in range(in_n_components):
    atributos.append('PC'+ str(i+1))
x_pca = pca.fit_transform(x_estandarizada)
pca_dataframe = pd.DataFrame(data = x_pca, columns=atributos)

x_fin = pca_dataframe
```

```python
explained_variance = pca.explained_variance_ratio_

print("\t\t\t Proporción de Explained Variance")
explained_variance_total = 0
i = 1
for element in explained_variance:
    print("PC" + str(i) + ":\t" + "{:.3f} |\t".format(element), end = '')
    if(i%5==0):
        print("\n")
    explained_variance_total += element
    i += 1

explained_variance_total *=100
print("TOTAL: " + "{:.2f}%".format(explained_variance_total))

plt.plot (pca.explained_variance_ratio_)
plt.ylabel('Explained Variance')
plt.xlabel('Componentes Principales')
plt.title('Explained Variance Ratio')
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Componente Principal 1', fontsize = 15)
ax.set_ylabel('Componente Principal  2', fontsize = 15)
ax.set_title('PCA de 2 componentes', fontsize = 20)

scatter = ax.scatter(x_pca[:,0],x_pca[:,1], c=data['clase'])

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Clases")
ax.add_artist(legend1)

ax.grid()
```

```python
data = [x_fin.PC1, x_fin.PC2]
fig7, ax7 = plt.subplots()
ax7.set_title('Varianza del PC1 y PC2')
ax7.boxplot(data)

print("\t     Detección de Outliers");
plt.show()
```

## Ensamble

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics
from mlxtend.classifier import EnsembleVoteClassifier


model_1 = KNeighborsClassifier(n_neighbors = in_n_neighbors)
model_2 = MLPClassifier(hidden_layer_sizes= in_hidden_layer_sizes,
                        solver='sgd',
                        max_iter=in_max_iter,
                        momentum=in_momentum)
model_3 = GaussianNB()

ensamble = EnsembleVoteClassifier(clfs=[model_1, model_2, model_3], weights=[1,1,1])
```

```python
scores_1 = cross_val_score(model_1, x_fin, y_fin, cv=in_splits)
scores_2 = cross_val_score(model_2, x_fin, y_fin, cv=in_splits)
scores_3 = cross_val_score(model_3, x_fin, y_fin, cv=in_splits)
scores_ensamble = cross_val_score(ensamble, x_fin, y_fin, cv=in_splits, scoring='accuracy')

print("\t -- Precisión tras cada k-Fold por clasificador --")
print("\t\tKNN\tNeural Net\tNaive Bayes\t[Ensamble]")
for i in range(in_splits):
    print("Pliegue " + str(i+1) + ":\t" + str(round(scores_1[i],2)) + "\t" + str(round(scores_2[i],2)) + "\t\t" + str(round(scores_3[i],2)) + "\t\t" + str(round(scores_ensamble[i],2)) + "\n" )

print("Media:\t\t%0.2f\t%0.2f\t\t%0.2f\t\t%0.2f" % (scores_1.mean(), scores_2.mean(), scores_3.mean(), scores_ensamble.mean()))
print("(+/-):\t\t%0.2f\t%0.2f\t\t%0.2f\t\t%0.2f" % (scores_1.std() * 2, scores_2.std() * 2, scores_3.std() * 2, scores_ensamble.std() * 2))
```

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec
import itertools

gs = gridspec.GridSpec(2, 2)

fig = plt.figure(figsize=(10,8))

labels = ['KNN', 'Neural Net', 'Naive Bayes', 'Ensemble']
for clf, lab, grd in zip([model_1, model_2, model_3, ensamble],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    ax = plt.subplot(gs[grd[0], grd[1]])
    X = x_fin.loc[:,['PC1', 'PC2']].values
    clf.fit(X, y_fin.to_numpy())
    fig = plot_decision_regions(X=X, y=y_fin.to_numpy(), clf=clf)
    plt.title(lab)
```

```python
for clf, lab in zip([model_1, model_2, model_3, ensamble], labels):
    y_pred = cross_val_predict(ensamble, X, y_fin, cv=in_splits)
    cm = confusion_matrix(y_fin, y_pred)
    cm_df = pd.DataFrame(cm)
    sns.heatmap(cm_df, annot=True, cmap="YlGnBu")
    plt.title('Matriz de confusión de ' + lab)
    plt.ylabel('verdadera')
    plt.xlabel('Predichas')
    plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(y, y_pred, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Reales')
ax.set_ylabel('Predichas')
plt.title('Clases reales vs Predecidas por el ensamble', loc='left')

plt.show()

ax.set_xlabel('Reales')
ax.set_ylabel('Predichas')
sns.kdeplot(y, y_pred,  cmap="Blues", shade=True)
plt.title('Gráfica de Densidad')
plt.show()
```
